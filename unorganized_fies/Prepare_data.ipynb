```python
#!/usr/bin/env python3
"""
extract_and_prepare_tensors.py

1) Reads Negative_organized.csv (sgRNA, off_site, chr, start, end, strand)
2) Extracts 8 epigenetic tracks from bigWig/bw files (NaN→0, reverse on '-')
3) One-hot encodes sgRNA and off_site (4 channels each, 23 bp)
4) Stacks [epi(8), sgRNA(4), off_site(4)] → (N, 16, 23)
5) Saves as PyTorch tensor (.pt)
"""

import os
import sys
import pandas as pd
import numpy as np
import torch

try:
    import pyBigWig
except ImportError:
    sys.exit("Error: pyBigWig not found. `pip install pyBigWig` first.")

# ─── CONFIG ────────────────────────────────────────────────────────────────────
bw_files = {
    'ATAC':    'ATAC.bigWig',
    'CTCF':    'CTCF_fold_change_over_control_hg19.bigWig',
    'H3K9me3': 'H3K9me3.bigWig',
    'H3K4me1': 'HEK293T_H3K4me1_ChIP.bw',
    'H3K27ac': 'hek293_h3k27ac.bw',
    'H3K4me3': 'hek293_h3k4me3.bw',
    'DNase':   'hg19_DNase_Hek293t.bigWig',
    'RRBS':  'merged.sorted.bw',
    'H3K27me3': 'H3K27me3.bigWig',
    'H3K36me3': 'H3K36me3.bigWig'
}

CSV_PATH = 'Negative_organized.csv'
OUT_PATH = 'Negative.pt'
# ────────────────────────────────────────────────────────────────────────────────

# 1) Load regions
if not os.path.exists(CSV_PATH):
    sys.exit(f"Error: {CSV_PATH} not found.")

df = pd.read_csv(CSV_PATH, index_col=0)
N = len(df)
# compute inclusive length
L = int(df.iloc[0]['end'] - df.iloc[0]['start'] + 1)

# 2) Extract epi signals into NumPy
epi = np.zeros((N, len(bw_files), L), dtype=np.float32)
for i, (_, row) in enumerate(df.iterrows()):
    chrom, start, end, strand = row['chr'], int(row['start']), int(row['end']), row['strand']
    if (end - start + 1) != L:
        sys.exit(f"Error: row {i+2} region length ≠ {L}")
    for j, path in enumerate(bw_files.values()):
        if not os.path.exists(path):
            sys.exit(f"Error: missing {path}")
        bw = pyBigWig.open(path)
        vals = np.array(bw.values(chrom, start, end+1), dtype=np.float32)
        bw.close()
        vals = np.nan_to_num(vals, nan=0.0)
        if strand == '-':
            vals = vals[::-1]
        epi[i, j, :] = vals

# 3) One-hot encode sgRNA / off_site
def one_hot(seqs, length):
    base2idx = {'A':0,'C':1,'G':2,'T':3}
    M = len(seqs)
    oh = torch.zeros((M, 4, length), dtype=torch.float32)
    for i, s in enumerate(seqs):
        s = s.upper()
        if len(s) != length:
            raise ValueError(f"Seq {i} len {len(s)} ≠ {length}")
        for k, nt in enumerate(s):
            idx = base2idx.get(nt)
            if idx is not None:
                oh[i, idx, k] = 1.0
    return oh

sg_list  = df['sgRNA'].tolist()
off_list = df['off_site'].tolist()

sg_oh  = one_hot(sg_list,  L)
off_oh = one_hot(off_list, L)

# 4) Stack: epi (8) + sgRNA (4) + off_site (4) → (N,16,L)
epi_t   = torch.from_numpy(epi)
stacked = torch.cat([epi_t, sg_oh, off_oh], dim=1)

# 5) Save
torch.save(stacked, OUT_PATH)
print(f"✅ Saved {OUT_PATH} with shape {tuple(stacked.shape)}")
```

```python
import torch
import matplotlib.pyplot as plt

def plot_tensor_sample(filepath, sample_idx=0):
    """
    Load a tensor file and plot a heatmap of the specified sample.

    Args:
        filepath (str): Path to the .pt file.
        sample_idx (int): Index of the sample to plot. Default is 0.
    """
    try:
        stacked = torch.load(filepath)
    except Exception as e:
        raise RuntimeError(f"Failed to load tensor from {filepath}: {e}")

    if sample_idx < 0 or sample_idx >= stacked.shape[0]:
        raise ValueError(f"sample_idx {sample_idx} is out of range (0 to {stacked.shape[0]-1})")

    sample = stacked[sample_idx]

    plt.figure(figsize=(10, 6))
    plt.imshow(sample.numpy(), aspect='auto', interpolation='nearest')
    plt.colorbar(label='Signal intensity')
    plt.title(f'Heatmap of sample {sample_idx}')
    plt.xlabel('Position (0 → {})'.format(sample.shape[1]-1))
    plt.ylabel('Channel (0 → {})'.format(sample.shape[0]-1))
    plt.tight_layout()
    plt.show()
plot_tensor_sample('Positive.pt', sample_idx=0)
plot_tensor_sample('Negative.pt', sample_idx=0)
```

```python
import torch

def compute_channel_mean_std(tensor, channel_indices):
    """Compute mean and std per channel for specified indices."""
    means = []
    stds = []
    for idx in channel_indices:
        data = tensor[:, idx, :]
        means.append(data.mean())
        stds.append(data.std())
    return torch.stack(means), torch.stack(stds)

def apply_z_score_normalization(tensor, channel_indices, means, stds):
    """Apply Z-score normalization using precomputed means and stds."""
    tensor = tensor.clone()
    for i, idx in enumerate(channel_indices):
        mean = means[i]
        std = stds[i]
        if std == 0:
            std = 1.0
        tensor[:, idx, :] = (tensor[:, idx, :] - mean) / std
    return tensor

```

```python
from matplotlib.colors import LogNorm
import torch
import torch
import matplotlib.pyplot as plt

def plot_normalized_heatmap(X, sample_idx=0, channels=range(18)):
    """
    Plot a heatmap of selected channels for a given sample from the normalized tensor.

    Args:
        X (torch.Tensor): Normalized input tensor of shape [N, 18, 23].
        sample_idx (int): Index of the sample to visualize.
        channels (list or range): Which channels to plot (default: all 18 channels).
    """
    # Channel labels
    channel_labels = [
        "ATAC-seq",
        "CTCF ChIP-seq",
        "H3K9me3 ChIP-seq",
        "H3K4me1 ChIP-seq",
        "H3K27ac ChIP-seq",
        "H3K4me3 ChIP-seq",
        "DNase-seq",
        "RRBS",
        "H3K27me3 ChIP-seq",
        "H3K36me3 ChIP-seq",
        "sgRNA_A",
        "sgRNA_C",
        "sgRNA_G",
        "sgRNA_T",
        "off-site_A",
        "off-site_C",
        "off-site_G",
        "off-site_T"
    ]
    
    # Safety checks
    if not isinstance(X, torch.Tensor):
        raise ValueError("Input X must be a torch.Tensor.")
    
    if sample_idx < 0 or sample_idx >= X.shape[0]:
        raise IndexError(f"Sample index {sample_idx} out of range. Total samples: {X.shape[0]}")

    sample = X[sample_idx, channels, :]  # Shape: [len(channels), 23]
    
    if sample.requires_grad:
        sample = sample.detach()
    if sample.is_cuda:
        sample = sample.cpu()
    
    # Prepare y-tick labels
    selected_labels = [channel_labels[i] for i in channels]

    # Plot
    plt.figure(figsize=(12, 6))
    plt.imshow(sample.numpy(), aspect='auto', interpolation='nearest')
    plt.colorbar(label='Z-score normalized signal')
    plt.title(f'Heatmap of Sample {sample_idx}')
    plt.xlabel('Sequence Position (0 → 22)')
    plt.yticks(ticks=range(len(channels)), labels=selected_labels)
    plt.tight_layout()
    plt.show()

def binary_stratified_random_split(
    X, y, test_ratio=0.2, test_chunk=1, seed=None
):
    """
    Random stratified split into train/test for binary labels.

    Args:
      X             – your feature tensor
      y             – your label tensor (0/1)
      test_ratio    – fraction for each fold (0.2 → 5 folds)
      test_chunk    – which fold to use as test (1-based index)
      seed          – for reproducibility

    Returns:
      X_train, X_test, y_train, y_test
    """
    # Basic checks
    if not (0 < test_ratio < 1):
        raise ValueError("test_ratio must be between 0 and 1")
    n_splits = int(1 / test_ratio)
    if not (1 <= test_chunk <= n_splits):
        raise ValueError(f"test_chunk must be in [1, {n_splits}]")

    # Set seed
    if seed is not None:
        torch.manual_seed(seed)

    # Prepare a fold‑assignment array
    fold_labels = torch.empty(len(y), dtype=torch.long)

    # Shuffle & assign folds separately for each class
    for cls in [0, 1]:
        cls_idx = (y == cls).nonzero(as_tuple=True)[0]
        permuted = cls_idx[torch.randperm(len(cls_idx))]
        base_size = len(permuted) // n_splits
        extras = len(permuted) % n_splits

        start = 0
        for fold in range(n_splits):
            size = base_size + (1 if fold < extras else 0)
            end = start + size
            fold_labels[permuted[start:end]] = fold
            start = end

    # Pick out train/test indices for the requested fold
    fold = test_chunk - 1
    test_idx  = (fold_labels == fold).nonzero(as_tuple=True)[0]
    train_idx = (fold_labels != fold).nonzero(as_tuple=True)[0]

    return X[train_idx], X[test_idx], y[train_idx], y[test_idx]

import os
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score, roc_curve, auc, precision_recall_curve, average_precision_score
from sklearn.model_selection import train_test_split
import seaborn as sns
for i in range(5):
    print(i+1)
    test_chunk_id=i+1

    # ====== Paths ======
    path_seq = '.'

    
    # ====== Load Combined Data ======
    combined_positive = torch.load(os.path.join(path_seq, 'Positive.pt')).float()
    combined_negative = torch.load(os.path.join(path_seq, 'Negative.pt')).float()
    
    # Create labels: 1 for positive, 0 for negative
    labels_positive = torch.ones(combined_positive.shape[0], dtype=torch.float32)
    labels_negative = torch.zeros(combined_negative.shape[0], dtype=torch.float32)
    
    # Concatenate data and labels
    X = torch.cat([combined_positive, combined_negative], dim=0)
    y = torch.cat([labels_positive, labels_negative], dim=0)
    
    # ====== Train-Test Split ======
    X_train, X_test, y_train, y_test = binary_stratified_random_split(X, y, test_ratio=0.2, test_chunk=test_chunk_id, seed=42)
    
    print(X_train.shape,y_train.shape)
    print(f"Train samples: {X_train.shape[0]}, Test samples: {X_test.shape[0]}")
    
    import torch
    
    # Define continuous channels
    continuous_channels = list(range(10))

    # Step 1: Compute means and stds *only on X_train*
    means, stds = compute_channel_mean_std(X_train, continuous_channels)
    
    # Step 2: Normalize both train and test using these statistics
    X_train_normalized = apply_z_score_normalization(X_train, continuous_channels, means, stds)
    X_test_normalized = apply_z_score_normalization(X_test, continuous_channels, means, stds)
    plot_normalized_heatmap(X_train_normalized, sample_idx=0)
    plot_normalized_heatmap(X_test_normalized, sample_idx=0)
    
    # ====== Balanced Batch Sampler ======
    def sample_balanced_batch(X_tr, y_tr, n_pos=400, n_neg=400):
        pos_indices = (y_tr == 1).nonzero(as_tuple=True)[0]
        neg_indices = (y_tr == 0).nonzero(as_tuple=True)[0]
        replace_pos = len(pos_indices) < n_pos
        replace_neg = len(neg_indices) < n_neg
        pos_sample_indices = np.random.choice(pos_indices.cpu().numpy(), n_pos, replace=replace_pos)
        neg_sample_indices = np.random.choice(neg_indices.cpu().numpy(), n_neg, replace=replace_neg)
        batch_X = torch.cat([X_tr[pos_sample_indices], X_tr[neg_sample_indices]], dim=0)
        batch_y = torch.cat([y_tr[pos_sample_indices], y_tr[neg_sample_indices]], dim=0)
        perm = torch.randperm(batch_X.size(0))
        return batch_X[perm], batch_y[perm]
    
    # ====== Transformer Model Definition ======
    class TransformerClassifier(nn.Module):
        def __init__(self, input_dim, seq_length, hidden_dim=256, num_layers=3, nhead=8, dropout=0.1):
            super(TransformerClassifier, self).__init__()
            self.embedding = nn.Linear(input_dim, hidden_dim)
            self.positional_encoding = nn.Parameter(torch.zeros(1, seq_length, hidden_dim))
            encoder_layer = nn.TransformerEncoderLayer(
                d_model=hidden_dim, nhead=nhead,
                dim_feedforward=hidden_dim*2, dropout=dropout, activation='relu'
            )
            self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
            self.layernorm = nn.LayerNorm(hidden_dim)
            self.fc = nn.Linear(hidden_dim, 1)
            self.sigmoid = nn.Sigmoid()

        def forward(self, x):
            x = x.transpose(1, 2)
            x = self.embedding(x) + self.positional_encoding
            x = self.layernorm(x)
            x = x.transpose(0, 1)
            x = self.transformer_encoder(x)
            x = x.transpose(0, 1).mean(dim=1)
            return self.sigmoid(self.fc(x)).squeeze()
    
    # ====== Initialize ======
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = TransformerClassifier(input_dim=18, seq_length=23).to(device)
    optimizer = optim.AdamW(model.parameters(), lr=1e-4, weight_decay=1e-2)
    
    num_epochs = 300
    train_losses = []
    test_losses = []
    train_accuracies = []
    test_accuracies = []
    fnr_history = []  # false negative rate history
    fpr_history = []  # false positive rate history
    recall_history = []   # recall history
    precision_history = []  # precision history
    best_fpr = float('inf')
    best_epoch = -1
    for epoch in range(num_epochs):
        model.train()
        inputs, labels = sample_balanced_batch(X_train_normalized, y_train, n_pos=400, n_neg=400)
        inputs, labels = inputs.to(device), labels.to(device)
    
        optimizer.zero_grad()
        outputs = model(inputs)  # [batch]
    
        # Compute weighted loss: assign higher weight to positive samples to reduce FNR.
        pos_weight = 2.0  # adjust as needed
        sample_weights = torch.where(labels == 1, torch.tensor(pos_weight, device=device), torch.tensor(1.0, device=device))
        loss = nn.functional.binary_cross_entropy(outputs, labels, weight=sample_weights)
    
        loss.backward()
        optimizer.step()
    
        train_losses.append(loss.item())
    
        # Compute training accuracy on this balanced batch
        preds = (outputs >= 0.5).float()
        train_acc = (preds == labels).float().mean().item()
        train_accuracies.append(train_acc)
    
        # ====== Evaluate on Test Set ======
        model.eval()
        test_loss = 0.0
        all_test_preds = []
        all_test_labels = []
        batch_size = 64
        with torch.no_grad():
            for i in range(0, X_test_normalized.shape[0], batch_size):
                batch_X = X_test_normalized[i:i+batch_size].to(device)
                batch_y = y_test[i:i+batch_size].to(device)
                batch_preds = model(batch_X)
                if batch_preds.ndim == 0:
                    batch_preds = batch_preds.unsqueeze(0)
                # Use standard BCE loss for evaluation (no weighting)
                loss_val = nn.functional.binary_cross_entropy(batch_preds, batch_y)
                test_loss += loss_val.item() * batch_X.size(0)
                all_test_preds.extend(batch_preds.cpu().numpy())
                all_test_labels.extend(batch_y.cpu().numpy())
        test_loss /= X_test_normalized.shape[0]
        test_losses.append(test_loss)
    
        # Compute test accuracy
        test_preds_binary = [1 if p >= 0.5 else 0 for p in all_test_preds]
        test_acc = accuracy_score(all_test_labels, test_preds_binary)
        test_accuracies.append(test_acc)

    # Compute FNR and FPR from test set confusion matrix
        cm = confusion_matrix(all_test_labels, test_preds_binary)
        if cm.size == 4:
            tn, fp, fn, tp = cm.ravel()
        else:
            # Handling edge cases where one of the classes might be missing
            tn = cm[0,0] if cm.shape[0]>0 and cm.shape[1]>0 else 0
            fp = cm[0,1] if cm.shape[1]>1 else 0
            fn = cm[1,0] if cm.shape[0]>1 else 0
            tp = cm[1,1] if cm.shape[0]>1 and cm.shape[1]>1 else 0
        fnr = fn / (fn + tp) if (fn + tp) > 0 else 0
        fpr = fp / (fp + tn) if (fp + tn) > 0 else 0
        fnr_history.append(fnr)
        fpr_history.append(fpr)
        if fnr < 0.1 and fpr < best_fpr:
            best_fpr = fpr
            best_epoch = epoch + 1
            torch.save({
                'epoch': best_epoch,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'fnr': fnr,
                'fpr': fpr
            }, f'best_model_{test_chunk_id}th.pth')
            print(f"👉 Saved new best model at epoch {best_epoch} (FNR={fnr:.4f}, FPR={fpr:.4f})")
        # ====== Compute Recall and Precision ======
        recall = tp / (tp + fn) if (tp + fn) > 0 else 0
        precision = tp / (tp + fp) if (tp + fp) > 0 else 0
        recall_history.append(recall)
        precision_history.append(precision)
    
        print(f"Epoch [{epoch+1}/{num_epochs}], Train Loss: {loss.item():.4f}, Train Acc: {train_acc:.4f}, Test Loss: {test_loss:.4f}, Test Acc: {test_acc:.4f}, FNR: {fnr:.4f}, FPR: {fpr:.4f}, Recall: {recall:.4f}, Precision: {precision:.4f}")
    print(f"\nTraining done — best model was from epoch {best_epoch} with FPR={best_fpr:.4f} (and FNR<0.1).")
    # ====== Plot Loss and Accuracy Curves ======
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    plt.plot(range(1, num_epochs+1), train_losses, label="Train Loss")
    plt.plot(range(1, num_epochs+1), test_losses, label="Test Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Loss Curve")
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(range(1, num_epochs+1), train_accuracies, label="Train Acc")
    plt.plot(range(1, num_epochs+1), test_accuracies, label="Test Acc")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.title("Accuracy Curve")
    plt.legend()
    plt.tight_layout()
    plt.show()
    
    # ====== Plot FNR and FPR Curves ======
    plt.figure(figsize=(6, 4))
    plt.plot(range(1, num_epochs+1), fnr_history, label="FNR")
    plt.plot(range(1, num_epochs+1), fpr_history, label="FPR")
    plt.xlabel("Epoch")
    plt.ylabel("Rate")
    plt.title("FNR and FPR Curves")
    plt.legend()
    plt.show()
    
    # ====== Plot Recall and Precision Curves ======
    plt.figure(figsize=(6, 4))
    plt.plot(range(1, num_epochs+1), recall_history, label="Recall")
    plt.plot(range(1, num_epochs+1), precision_history, label="Precision")
    plt.xlabel("Epoch")
    plt.ylabel("Rate")
    plt.title("Recall and Precision Curves")
    plt.legend()
    plt.show()
    # ===== Add inside the fold loop AFTER computing all_test_preds, all_test_labels, and test_preds_binary =====

    # ROC Curve
    fpr_curve, tpr_curve, _ = roc_curve(all_test_labels, all_test_preds)
    roc_auc = auc(fpr_curve, tpr_curve)
    plt.figure()
    plt.plot(fpr_curve, tpr_curve, label=f'Fold {test_chunk_id} (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], '--', linewidth=0.8)
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve - Fold {test_chunk_id}')
    plt.legend(loc='lower right')
    plt.tight_layout()
    plt.show()
    
    # Precision–Recall Curve
    precision_curve, recall_curve, _ = precision_recall_curve(all_test_labels, all_test_preds)
    pr_auc = average_precision_score(all_test_labels, all_test_preds)
    plt.figure()
    plt.plot(recall_curve, precision_curve, label=f'Fold {test_chunk_id} (AUC = {pr_auc:.2f})')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title(f'Precision–Recall Curve - Fold {test_chunk_id}')
    plt.legend(loc='lower left')
    plt.tight_layout()
    plt.show()

    # Confusion Matrix Heatmap
    cm = confusion_matrix(all_test_labels, test_preds_binary)
    plt.figure(figsize=(5, 4))
    sns.heatmap(
        cm,
        annot=True,
        fmt='d',
        cmap='Reds',
        norm=LogNorm(vmin=max(cm.min(), 1), vmax=cm.max()),
        xticklabels=['Predicted Neg', 'Predicted Pos'],
        yticklabels=['Actual Neg', 'Actual Pos']
    )
    plt.title(f'Confusion Matrix - Fold {test_chunk_id}')
    plt.ylabel('Actual Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.show()

    
```

```python
import os
import torch
import torch.nn as nn
import numpy as np
from sklearn.metrics import (
    confusion_matrix,
    roc_curve, auc,
    precision_recall_curve, average_precision_score
)
import matplotlib.pyplot as plt
import seaborn as sns

# ——— Constants ———
continuous_channels = list(range(10))
batch_size = 64

# ——— Define your exact same model architecture ———
class TransformerClassifier(nn.Module):
    def __init__(self, input_dim=18, seq_length=23, hidden_dim=256, num_layers=3, nhead=8, dropout=0.1):
        super(TransformerClassifier, self).__init__()
        self.embedding = nn.Linear(input_dim, hidden_dim)
        self.positional_encoding = nn.Parameter(torch.zeros(1, seq_length, hidden_dim))
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=hidden_dim, nhead=nhead,
            dim_feedforward=hidden_dim*2, dropout=dropout, activation='relu'
        )
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        self.layernorm = nn.LayerNorm(hidden_dim)
        self.fc = nn.Linear(hidden_dim, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = x.transpose(1, 2)
        x = self.embedding(x) + self.positional_encoding
        x = self.layernorm(x)
        x = x.transpose(0, 1)
        x = self.transformer_encoder(x)
        x = x.transpose(0, 1).mean(dim=1)
        return self.sigmoid(self.fc(x)).squeeze()

# ——— Load Data ———
path = '.'
pos = torch.load(os.path.join(path, 'Positive.pt')).float()
neg = torch.load(os.path.join(path, 'Negative.pt')).float()
X = torch.cat([pos, neg], dim=0)
y = torch.cat([torch.ones(len(pos)), torch.zeros(len(neg))], dim=0).numpy()

# Normalization
def compute_channel_mean_std(X, channels):
    means = X[:, channels, :].mean(dim=(0, 2))
    stds = X[:, channels, :].std(dim=(0, 2)) + 1e-8
    return means, stds

def apply_z_score_normalization(X, channels, means, stds):
    X_norm = X.clone()
    for i, ch in enumerate(channels):
        X_norm[:, ch, :] = (X[:, ch, :] - means[i]) / stds[i]
    return X_norm

means, stds = compute_channel_mean_std(X, continuous_channels)
print(means,stds)
X = apply_z_score_normalization(X, continuous_channels, means, stds)

# ——— Load the 5 models ———
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
models = []
for fold in range(1, 6):
    m = TransformerClassifier().to(device)
    ckpt = torch.load(f'best_model_{fold}th.pth', map_location=device)
    m.load_state_dict(ckpt['model_state_dict'])
    m.eval()
    models.append(m)

# ——— Inference: collect each model’s scores in batches ———
with torch.no_grad():
    all_scores = []
    for m in models:
        preds = []
        for i in range(0, X.shape[0], batch_size):
            batch = X[i:i+batch_size].to(device)
            pred = m(batch).cpu().numpy()
            preds.append(pred)
        preds = np.concatenate(preds, axis=0)
        all_scores.append(preds)
    all_scores = np.stack(all_scores, axis=0)

# ——— Ensemble by averaging ———
ensemble_scores = all_scores.mean(axis=0)
y_pred = (ensemble_scores >= 0.5).astype(int)

# ——— 1. Confusion Matrix ———
cm = confusion_matrix(y, y_pred)
cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

# Plot
plt.figure(figsize=(5, 4))
sns.heatmap(cm_normalized, annot=cm, fmt='d', cmap='Blues')
plt.title("Row-Normalized Confusion Matrix (Heat Separated)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# ——— 2. ROC Curve & AUC ———
fpr, tpr, _ = roc_curve(y, ensemble_scores)
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(6,4))
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.3f}")
plt.plot([0,1], [0,1], '--', alpha=0.5)
plt.title("Ensemble ROC Curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend(loc="lower right")
plt.show()

# ——— 3. Precision–Recall Curve & AP ———
precision, recall, _ = precision_recall_curve(y, ensemble_scores)
ap = average_precision_score(y, ensemble_scores)
plt.figure(figsize=(6,4))
plt.plot(recall, precision, label=f"AP = {ap:.3f}")
plt.title("Ensemble Precision–Recall Curve")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.legend(loc="upper right")
plt.show()
```
